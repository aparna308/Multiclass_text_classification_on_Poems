```python
import nltk
import pandas as pd
```

```python
df=pd.read_csv('Poem_classification_dataset')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
#Removing duplicate rows
df = df.drop_duplicates()
```

```python
# Shuffle the Dataset
df = df.sample(frac=1,random_state=4)
```

```python
df.Genre.value_counts()
#Dataset is balanced so no need of over sampling or undersampling.
```

```python
df['Genre']=df['Genre'].map({'Music':0,'Death':1,'Environment':2,'Affection':3})
```

```python
df['Genre'] = df['Genre'].astype(int)
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
x=df['Poem']
```

```python
x=x.str.lower()
```

```python
import re
x.replace('[^ a-zA-Z]',' ',regex=True,inplace=True)
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
for i in x:
    words=i.split()
    words=[stemmer.stem(word) for word in words if not word in set(stopwords.words('english'))]
    i=' '.join(words)
    print(i)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfobj=TfidfVectorizer()
x_tfidf=tfidfobj.fit_transform(x).toarray()
x_tfidf
```

```python
x_tfidf.shape
```

```python
y=df['Genre']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
x_tfidf,y,test_size=0.3,random_state=0)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
random_search=RandomizedSearchCV(NBmodel,param_distributions=params,n_iter=5,scoring='roc_auc',cv=5,verbose=1)
random_search.fit(x_train,y_train)
random_search.best_estimator_
```
